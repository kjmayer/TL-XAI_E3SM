```python
%%time
def preprocess(ds):
    return ds.sel(level=500, latitude = slice(90,20))

era5_path = '/gpfs/csfs1/collections/rda/data/ds633.0/e5.oper.an.pl/*/' # ERA5 data 
era5_save_path = '/glade/derecho/scratch/kjmayer/DATA/ERA5/z500/daily/'
filenames = glob.glob(era5_path+'e5.oper.an.pl.128_129_z.ll025sc.*.nc')
filenames.sort()

for f,file in enumerate(filenames[24946:-305]): #[20454:-305]: 1996.01.01-2023.12.31 #[14245:-305]: 1979-2023
    if os.path.exists(era5_save_path+file.split('/')[-1][:-16]+'.nc'):
        print("File exists. Skipping.")
    else:
        z500 = xr.open_mfdataset(file,preprocess=preprocess)
        tempday = z500['Z'][0].time
        z500_mean = z500.mean('time')
        z500_mean = z500_mean.assign_coords(time=tempday)
    
        z500_mean.to_netcdf(era5_save_path+file.split('/')[-1][:-16]+'.nc')
        del z500, z500_mean
```

```python
gpcp_path = '/gpfs/csfs1/collections/rda/data/ds728.7/gpcp_v1.3_daily/' # GPCP data 
gpcp_save_path = '/glade/derecho/scratch/kjmayer/DATA/GPCP/PRECT/daily/'

years = np.arange(1996,2024,1)
for year in years:
    print(year)
    gpcp = xr.open_mfdataset(gpcp_path+str(year)+'/*.nc',
                             concat_dim = 'time',
                             combine="nested",
                             preprocess = lambda ds: (ds['precip'].where((ds['precip'].latitude >= -30) & (ds['precip'].latitude <= 30), drop = True))
                            )
    gpcp = gpcp.sortby('time')
    gpcp.to_netcdf(gpcp_save_path+'precip_30S-30N_'+str(np.unique(gpcp.time.dt.year)[0])+'.nc')
```
